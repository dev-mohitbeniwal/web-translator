```python
import os
from fnmatch import fnmatch
from googletrans import Translator
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize


def translate_to_hindi(text):
    translator = Translator()
    translation = translator.translate(text, src='en', dest='hi')
    return translation.text

# Function to replace English text with Hindi text in a single HTML file
def replace_text_in_file(html_file_path):
    with open(html_file_path, 'r',encoding="utf8") as file:
        html_content = file.read()

    # Find all the text elements in the HTML file
#     text_elements = soup.findAll(text=True)
#     print(text_elements)
    
    soup = BeautifulSoup(html_content, 'html.parser')
    for script in soup(["script", "style"]):
        script.extract()    # rip it out

    # get text
    text = soup.get_text()
    print(text)

#     # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
#     # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
#     # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)





    hindi_text = translate_to_hindi(text)
    text.replace_with(hindi_text)

    # Write the translated HTML content to a new file
    new_html_file_path = os.path.splitext(html_file_path)[0] + "_hi.html"
    with open(new_html_file_path, 'w') as file:
        file.write(str(soup))

    

# Function to parse a directory and replace text in all HTML files
def replace_text_in_dir(root,pattern,l):
    for path, subdirs, files in os.walk(root):
        for name in files:
            if fnmatch(name, pattern):
                l.append(os.path.join(path, name))
                
                
    

# Example usage:
l=list()
pattern = "*.html"
replace_text_in_dir("C:\My Web Sites\classcentral_clone",pattern,l)
for path in l:
    replace_text_in_file(path)
                         
```
